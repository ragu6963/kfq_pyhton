```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse
url = "https://news.naver.com"
```

```python
# web_url에 원하는 웹의 URL을 넣어주시면 됩니다.
with urllib.request.urlopen(url+"/main/home.nhn") as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup)
```

```python
news = soup.select(".section_list_ranking")
```

```python
news
```

```python

news_list = []
for n in news:
    links = n.select("a")
    temp = []
    for link in links:
        href = link['href']
        title = link.get_text()
        temp.append({"link":href,"title":title})
    news_list.append(temp)
    
news_list[0]
```

```python
link = news_list[0][0]['link']
with urllib.request.urlopen(url+link) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser') 
```

```python
content = soup.select_one('._article_body_contents')
print(content.get_text())
```

```python
news_result_list = []

for news in news_list:
    for n in news:
        new_url = url + n['link']
        with urllib.request.urlopen(new_url) as response:
            html = response.read()
            soup = BeautifulSoup(html, 'html.parser')
        content = soup.select_one('._article_body_contents').get_text()
        content = content.replace("// flash 오류를 우회하기 위한 함수 추가","")
        content = content.replace("function _flash_removeCallback() {}","")
        content = content.replace("\n","")
        content = content.strip()
        news_result_list.append({"title":n['title'],"url":new_url,"content":content})
 
```

```python
print("제목 : {} \n링크 : {} \n\n내용 \n {}".format(news_result_list[0]['title'],news_result_list[0]['url'],news_result_list[0]['content']))
```

```python
for news in news_result_list:
    print("제목 : {}\n링크 : {} ".format(news['title'],news['url']))
```
