```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse
url = "https://music.bugs.co.kr/chart"
```

```python
# web_url에 원하는 웹의 URL을 넣어주시면 됩니다.
with urllib.request.urlopen(url) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup)
```

```python
trs = soup.select("table.list > tbody >tr ")
```

```python
list = []
```

```python
for tr in trs:
    ranking = tr.select('.ranking >strong')[0].get_text()
    music = tr.select('p.title >a')[0].get_text()
    artist = tr.select('p.artist >a')[0].get_text()
    list.append((ranking,music,artist))
    print("{}등. {} - {}".format(ranking,music,artist))    
```
