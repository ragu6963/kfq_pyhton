```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse
url = "https://finance.naver.com/marketindex/?tabSel=exchange#tab_section"
```

```python
# web_url에 원하는 웹의 URL을 넣어주시면 됩니다.
with urllib.request.urlopen(url) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.prettify()
```

```python
table = soup.select('.view > .selectbox-noscript > .selectbox-source >option')
```

```python
print(table)
```

```python
for t in table:
    print(t)
```

```python
for t in table:
    print("{:<20} - {:>5}원".format(t.get_text(),t["value"]))
    
```
